```python
 #creating passenger obj with response of 1 passenger group
class help_passengers:
    def __init__(self,response):
        self.response=response
        self.yes=0
        self.unique_response=[]
        self.questions_list=[]
        self.questions_list_common=[]
        self.questions_list_notcommon=[]
    
    def count_yes_all(self):
        for line in self.response.split("\n"):
            print(line)
            for question in line:
                if question not in self.questions_list:
                    self.questions_list.append(question)
        return len(self.questions_list)
    
    #fn to give no. of unique questions that everyone in the group said yes to
    def count_yes_common(self):
        #getting first passenger's response
        psngr1_questions=(self.response.split("\n")[:1])[0]
        
        #creating list with alphabets in first passenger's response
        for question in psngr1_questions:
            self.questions_list_common.append(question)
        print("list of possible questions ",self.questions_list_common)
        print()
        
        #checking if letters in first passenger's response are in other passengers' responses
        #if length of not common list is >= skip loop
        for line in self.response.split("\n")[1:]:
            if len(set(self.questions_list_common))>len(set(self.questions_list_notcommon)):
                print("passengers response ",line)
                
                for index,question in enumerate(self.questions_list_common):
                    question_check=self.questions_list_common[index]
                    print("checking for ",question_check)
                    
                    
                    if question not in line:
                        print("removing ",question_check)
                        self.questions_list_notcommon.append(question_check)
                        print("list of not common questions ",self.questions_list_notcommon)
                    
                    elif question in line:
                        print(question," is in ",line)
            else:
                break
    
        val=len(set(self.questions_list_common))-len(set(self.questions_list_notcommon))
        print('no. of common answers',val)

        return val
```

```python
start = time.process_time()
total_yes_all=0
total_yes_common=0
for line in data_str.split("\n\n"):
    x=help_passengers(line)
    psngr_yes_common=x.count_yes_common()
    total_yes_common+=psngr_yes_common
```
